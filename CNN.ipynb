```python
# 1. Import Library
import random  # Untuk operasi terkait angka acak (misalnya, shuffling data).
import os  # Untuk operasi sistem seperti pengelolaan file dan direktori.
import numpy as np  # Untuk operasi matematika dan manipulasi array numerik.
import pandas as pd  # Untuk manipulasi dan analisis data berbasis tabel (dataframe).
import matplotlib.pyplot as plt  # Untuk membuat visualisasi data seperti grafik dan plot.
import seaborn as sns  # Untuk membuat visualisasi data yang lebih menarik dan informatif dibandingkan matplotlib.

# 2. Import library untuk evaluasi model
from sklearn.metrics import confusion_matrix, classification_report
# confusion_matrix: Untuk membuat matriks kebingungan (confusion matrix) guna mengevaluasi performa model.
# classification_report: Untuk menghasilkan laporan evaluasi model seperti precision, recall, F1-score, dll.

# 3. Import library dari Keras untuk membuat model neural network
from keras.models import Sequential  # Untuk membangun model neural network secara bertahap (layer by layer).
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout, BatchNormalization
# Conv2D: Layer konvolusi untuk ekstraksi fitur dari gambar.
# MaxPooling2D: Pooling layer untuk mengurangi dimensi data setelah konvolusi.
# Dense: Layer fully connected (lapisan neuron penuh) untuk keputusan klasifikasi.
# Flatten: Layer untuk meratakan data dari dimensi tinggi menjadi vektor 1D.
# Dropout: Layer regulasi untuk mengurangi overfitting dengan "menghapus" neuron secara acak selama pelatihan.
# BatchNormalization: Layer untuk normalisasi data selama pelatihan agar stabil dan mempercepat pelatihan.

# 4. Import library dari TensorFlow untuk augmentasi gambar
from tensorflow.keras.preprocessing.image import ImageDataGenerator
# ImageDataGenerator: Untuk membuat generator augmentasi gambar seperti rotasi, flipping, zoom, dll.

# 5. Import library untuk callback selama pelatihan model
from keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
# ModelCheckpoint: Untuk menyimpan model terbaik selama pelatihan berdasarkan metrik tertentu.
# EarlyStopping: Untuk menghentikan pelatihan lebih awal jika performa model tidak meningkat.
# ReduceLROnPlateau: Untuk mengurangi laju pembelajaran (learning rate) jika performa model stagnan.

# 6. Import optimizer dari Keras
from keras.optimizers import Adam  # Optimizer Adam untuk memperbarui bobot model selama pelatihan.

# 7. Magic command untuk menampilkan plot secara langsung dalam notebook
%matplotlib inline  # (Hanya untuk Jupyter Notebook) Menampilkan plot langsung di dalam notebook.
```

```python
# 1.Lokasi direktori train dan test
train_dir = r"E:\Pengenalan Wajah CNN K5\train"
test_dir = r"E:\Pengenalan Wajah CNN K5\test"

# Konfigurasi jumlah file yang diinginkan
required_train_count = 3200  # Per kelas untuk training
required_test_count = 800    # Per kelas untuk testing

def get_file_count_and_files(directory):
    """Mengembalikan jumlah file dan daftar file dalam folder."""
    files = [os.path.join(directory, f) for f in os.listdir(directory) if os.path.isfile(os.path.join(directory, f))]
    return len(files), files

def adjust_files(directory, required_count):
    """Menghapus file jika lebih dari jumlah yang dibutuhkan."""
    current_count, files = get_file_count_and_files(directory)
    if current_count > required_count:
        # Randomly delete files to meet the requirement
        files_to_delete = random.sample(files, current_count - required_count)
        for file in files_to_delete:
            os.remove(file)
    return current_count, len(files_to_delete) if current_count > required_count else 0

def process_directory(base_dir, required_count):
    """Proses setiap subfolder untuk memeriksa jumlah file."""
    total_before = 0
    total_removed = 0
    for class_dir in os.listdir(base_dir):
        class_path = os.path.join(base_dir, class_dir)
        if os.path.isdir(class_path):
            before, removed = adjust_files(class_path, required_count)
            total_before += before
            total_removed += removed
            print(f"Kelas: {class_dir}, Sebelum: {before}, Dihapus: {removed}")
    total_after = total_before - total_removed
    print(f"Total Sebelum: {total_before}, Total Sesudah: {total_after}")
    return total_before, total_after

print("Memproses data training...")
train_before, train_after = process_directory(train_dir, required_train_count)

print("\nMemproses data testing...")
test_before, test_after = process_directory(test_dir, required_test_count)
```

```python
# 1.1 Fungsi untuk Menampilkan Informasi Dataset
def display_dataset_info(train_dir, test_dir):
    print("\n Informasi Dataset \n")
    
    # Informasi Jumlah Data Training
    print("Jumlah Data Training:")
    total_train_images = 0
    for emotion in sorted(os.listdir(train_dir)):
        emotion_path = os.path.join(train_dir, emotion)
        if os.path.isdir(emotion_path):
            num_images = len(os.listdir(emotion_path))
            total_train_images += num_images
            print(f"{emotion}: {num_images} images")
    print(f"Total Training Images: {total_train_images}")
    
    # Informasi Jumlah Data Testing
    print("\nJumlah Data Testing:")
    total_test_images = 0
    for emotion in sorted(os.listdir(test_dir)):
        emotion_path = os.path.join(test_dir, emotion)
        if os.path.isdir(emotion_path):
            num_images = len(os.listdir(emotion_path))
            total_test_images += num_images
            print(f"{emotion}: {num_images} images")
    print(f"Total Testing Images: {total_test_images}")

# Panggil fungsi dengan direktori Anda
train_dir = 'E:/Pengenalan Wajah CNN K5/train'
test_dir = 'E:/Pengenalan Wajah CNN K5/test'

display_dataset_info(train_dir, test_dir)
```

```python
# 1.2. Membuat Folder untuk Menyimpan Hasil
print("2. Membuat Folder Hasil")
os.makedirs('hasil_visual', exist_ok=True) 
os.makedirs('hasil_csv', exist_ok=True)
print("✓ Folder hasil_visual dan hasil_csv berhasil dibuat\n")
```

```python
# 2. Konfigurasi Parameter
IMG_SIZE = 48
BATCH_SIZE = 64  # Ditingkatkan untuk training lebih cepat
EPOCHS = 50     # Ditambah epoch dengan early stopping
train_dir = 'E:/Pengenalan Wajah CNN K5/train'
test_dir = 'E:/Pengenalan Wajah CNN K5/test'
```

```python
# 3. Data Generator dengan Augmentasi
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range=0.3,
    height_shift_range=0.3,
    horizontal_flip=True,
    vertical_flip=False,
    fill_mode='nearest',
    zoom_range=0.2,
    shear_range=0.15,
    brightness_range=[0.8,1.2]
)

test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    color_mode='grayscale',
    shuffle=True
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    color_mode='grayscale',
    shuffle=False
)
```

```python
# 3.1 Visualisasi Sampel Data
print("\nMenampilkan 1 Sampel per Kelas...")
plt.figure(figsize=(15, 8))

# Dictionary untuk memastikan hanya 1 sampel per kelas
sampled_classes = {}

# Ambil batch dari train_generator hingga semua kelas memiliki 1 sampel
while len(sampled_classes) < len(train_generator.class_indices):
    batch = next(train_generator)
    images = batch[0]
    labels = batch[1]

    for image, label in zip(images, labels):
        class_index = np.argmax(label)
        class_name = list(train_generator.class_indices.keys())[class_index]

        # Jika kelas belum diambil, tambahkan sampel
        if class_name not in sampled_classes:
            sampled_classes[class_name] = image

            # Hentikan jika semua kelas telah diambil
            if len(sampled_classes) == len(train_generator.class_indices):
                break

# Visualisasikan sampel data
for i, (class_name, image) in enumerate(sampled_classes.items()):
    plt.subplot(1, len(sampled_classes), i + 1)
    plt.imshow(image.reshape(IMG_SIZE, IMG_SIZE), cmap='gray')
    plt.title(f'Class: {class_name}')
    plt.axis('off')

plt.suptitle('1 Sampel per Kelas', y=1.05)
plt.tight_layout()

# Menyimpan visualisasi ke folder hasil_visual
visual_path = os.path.join('hasil_visual', 'sampel_per_kelas_training.png')
plt.savefig(visual_path)
print(f"✓ Visualisasi 1 sampel per kelas disimpan di {visual_path}")

plt.show()
```

```python
# 3.2 Analisis Distribusi Kelas
print("\nDistribusi Kelas Dataset:")
class_counts = {}
for class_name in os.listdir(train_dir):
    class_counts[class_name] = len(os.listdir(os.path.join(train_dir, class_name)))

plt.figure(figsize=(10, 6))
sns.barplot(x=list(class_counts.keys()), y=list(class_counts.values()))
plt.title('Distribusi Kelas Dataset Training')
plt.xlabel('Kelas')
plt.ylabel('Jumlah Gambar')
plt.xticks(rotation=45)
plt.tight_layout()

# Menyimpan visualisasi ke folder hasil_visual
visual_path = os.path.join('hasil_visual', 'distribusi_kelas_training.png')
plt.savefig(visual_path)
print(f"✓ Visualisasi distribusi kelas dataset disimpan di {visual_path}")

plt.show()

for class_name, count in class_counts.items():
    print(f"{class_name}: {count} gambar")
```

```python
import pandas as pd

# 3.3 Analisis Normalisasi Data
print("\nData setelah normalisasi:")
x_batch, _ = next(train_generator)

# Menyimpan semua nilai pixel ke dalam DataFrame
pixel_data = pd.DataFrame(x_batch[0, :, :, 0].flatten(), columns=['Pixel_Value'])

# Menampilkan hanya 15 sampel nilai pixel di terminal
for i in range(15):
    print(f"[ {pixel_data.iloc[i, 0]:.8f} ]")

# Menyimpan semua data nilai pixel ke file CSV
csv_path = os.path.join('hasil_csv', 'normalisasi_pixel_values.csv')
pixel_data.to_csv(csv_path, index=False)
print(f"\n✓ Data nilai pixel setelah normalisasi disimpan di {csv_path}")

# Visualisasi distribusi nilai pixel
plt.figure(figsize=(10, 6))
sns.histplot(data=pixel_data['Pixel_Value'], bins=50)
plt.title('Distribusi Nilai Pixel Setelah Normalisasi')
plt.xlabel('Nilai Pixel')
plt.ylabel('Frekuensi')

# Menyimpan visualisasi ke folder hasil_visual
visual_path = os.path.join('hasil_visual', 'distribusi_pixel_normalisasi.png')
plt.savefig(visual_path)
print(f"✓ Visualisasi distribusi nilai pixel disimpan di {visual_path}")

plt.show()
```

```python
# 3.5 Analisis Ukuran Data
print("\nAnalisis Ukuran Data:")
print(f"Ukuran Input Gambar: {IMG_SIZE}x{IMG_SIZE} pixels")
print(f"Batch Size: {BATCH_SIZE}")
print(f"Steps per Epoch: {len(train_generator)}")
print(f"Total Training Samples: {len(train_generator.classes)}")
print(f"Total Testing Samples: {len(test_generator.classes)}")
```

```python
# 4. Model CNN 
print("\n4. Membangun Model CNN yang Dioptimalkan")
model = Sequential([
    # Block 1
    Conv2D(64, (3, 3), padding='same', activation='relu', input_shape=(IMG_SIZE, IMG_SIZE, 1)),
    Conv2D(64, (3, 3), padding='same', activation='relu'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),
    Dropout(0.3),
    
    # Block 2
    Conv2D(128, (3, 3), padding='same', activation='relu'),
    Conv2D(128, (3, 3), padding='same', activation='relu'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),
    Dropout(0.3),
    
    # Block 3
    Conv2D(256, (3, 3), padding='same', activation='relu'),
    Conv2D(256, (3, 3), padding='same', activation='relu'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),
    Dropout(0.4),
    
    # Dense Layers
    Flatten(),
    Dense(512, activation='relu'),
    BatchNormalization(),
    Dropout(0.5),
    Dense(256, activation='relu'),
    BatchNormalization(),
    Dropout(0.5),
    Dense(5, activation='softmax')
])

# Print Model Summary
print("\nArsitektur Model CNN:")
model.summary()
```

```python
# 5. Kompilasi Model
initial_learning_rate = 0.001
optimizer = Adam(learning_rate=initial_learning_rate)

model.compile(
    optimizer=optimizer,
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# 6. Callbacks
checkpoint = ModelCheckpoint(
    'E:/Pengenalan Wajah CNN K5/best_model.keras',
    monitor='val_accuracy',
    mode='max',
    save_best_only=True,
    verbose=1
)

early_stopping = EarlyStopping(
    monitor='val_accuracy',
    patience=15,
    restore_best_weights=True,
    verbose=1
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.2,
    patience=5,
    min_lr=1e-6,
    verbose=1
)
```

```python
# 7. Training
print("\nMemulai Training...")
history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=test_generator,
    callbacks=[checkpoint, early_stopping, reduce_lr]
)
```

```python
# Analisis Loss per Epoch
print("\nNilai Kesalahan (loss) per Epoch:")

# Menyimpan data loss ke dalam DataFrame
loss_data = pd.DataFrame({'Epoch': range(len(history.history['loss'])), 
                          'Loss': history.history['loss']})

# Menampilkan hanya 15 data loss pertama (setiap 10 epoch) di terminal
print("15 Sampel Nilai Loss:")
for epoch in range(0, min(50, len(history.history['loss'])), 10):
    loss = history.history['loss'][epoch]
    print(f"Epoch {epoch:3d}, Loss: {loss:.15f}")

# Tambahkan epoch ke-49 secara manual jika tersedia
if len(history.history['loss']) > 49:
    loss = history.history['loss'][49]
    print(f"Epoch  49, Loss: {loss:.15f}")

# Menyimpan semua data loss ke file CSV
csv_path = os.path.join('hasil_csv', 'loss_per_epoch.csv')
loss_data.to_csv(csv_path, index=False)

# Visualisasi Loss
plt.figure(figsize=(10, 6))
plt.plot(loss_data['Epoch'], loss_data['Loss'], label='Loss', color='blue')
plt.title('Nilai Kesalahan (loss) per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()

# Menyimpan visualisasi ke folder hasil_visual
visual_path = os.path.join('hasil_visual', 'loss_per_epoch.png')
plt.savefig(visual_path)

plt.show()
```

```python
# 8. Evaluasi dan Visualisasi
test_loss, test_accuracy = model.evaluate(test_generator)
print(f'\nTest accuracy: {test_accuracy:.4f}')

# Simpan hasil evaluasi ke CSV
evaluation_data = pd.DataFrame({'Metric': ['Test Loss', 'Test Accuracy'], 
                                'Value': [test_loss, test_accuracy]})
evaluation_csv_path = os.path.join('hasil_csv', 'model_evaluation.csv')
evaluation_data.to_csv(evaluation_csv_path, index=False)

# Prediksi dan data sebenarnya
predictions = model.predict(test_generator)
y_pred = np.argmax(predictions, axis=1)
y_true = test_generator.classes

# Simpan prediksi dan nilai sebenarnya ke CSV
prediction_data = pd.DataFrame({'True Label': y_true, 'Predicted Label': y_pred})
predictions_csv_path = os.path.join('hasil_csv', 'predictions.csv')
prediction_data.to_csv(predictions_csv_path, index=False)

# Plot training history
plt.figure(figsize=(15, 5))

# Plot Akurasi
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy', color='blue')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', color='orange')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Plot Loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss', color='blue')
plt.plot(history.history['val_loss'], label='Validation Loss', color='orange')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# Simpan visualisasi history ke file
visual_path = os.path.join('hasil_visual', 'training_history.png')
plt.savefig(visual_path)

plt.tight_layout()
plt.show()
```

```python
# 8. Evaluasi Model
print("\nClassification Report:")

# Mendapatkan prediksi
Y_pred = model.predict(test_generator)
y_pred = np.argmax(Y_pred, axis=1)
y_true = test_generator.classes

# Mendapatkan nama kelas
class_labels = list(test_generator.class_indices.keys())

# Membuat classification report
report = classification_report(
    y_true, 
    y_pred, 
    target_names=class_labels, 
    output_dict=True,
    digits=6  # Mengatur presisi desimal
)

# Mengkonversi report ke DataFrame
df_report = pd.DataFrame(report).transpose()

# Mengatur format tampilan
pd.set_option('display.float_format', lambda x: '%.6f' % x)
df_report = df_report.round(6)

# Mengatur ulang nama kolom
df_report.columns = ['precision', 'recall', 'f1-score', 'support']

# Menampilkan hasil di terminal
print(df_report)

# Menyimpan Classification Report ke CSV
csv_path = os.path.join('hasil_csv', 'classification_report.csv')
df_report.to_csv(csv_path, index=True)
```

```python
# ROC Curve dan AUC (Receiver Operating Characteristic)
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize

# One-hot encode y_true
y_true_binarized = label_binarize(y_true, classes=range(len(class_labels)))

# Plot ROC curve for each class
plt.figure(figsize=(10, 8))
for i, class_name in enumerate(class_labels):
    fpr, tpr, _ = roc_curve(y_true_binarized[:, i], Y_pred[:, i])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'Class {class_name} (AUC = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.title('ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.grid()
plt.tight_layout()
plt.savefig(os.path.join('hasil_visual', 'roc_curve.png'))
plt.show()
```

```python
# Simulasikan learning rate berdasarkan ReduceLROnPlateau
initial_learning_rate = 0.001  # Sama seperti yang digunakan dalam model.compile
factor = 0.2  # Nilai ReduceLROnPlateau
patience = 5  # Epoch tanpa perbaikan sebelum learning rate diturunkan

loss_values = history.history['loss']  # Ambil nilai loss
learning_rates = [initial_learning_rate]  # Mulai dengan learning rate awal

# Simulasikan penurunan learning rate
no_improve_epochs = 0
for i in range(1, len(loss_values)):
    if loss_values[i] >= loss_values[i-1]:  # Tidak ada perbaikan
        no_improve_epochs += 1
    else:  # Ada perbaikan
        no_improve_epochs = 0

    if no_improve_epochs > patience:  # Turunkan learning rate
        learning_rates.append(learning_rates[-1] * factor)
        no_improve_epochs = 0
    else:
        learning_rates.append(learning_rates[-1])

# Plot Learning Rate vs Loss
plt.figure(figsize=(10, 6))
plt.plot(learning_rates, loss_values, marker='o', label='Training Loss')
plt.xscale('log')  # Scale log untuk visualisasi learning rate
plt.title('Learning Rate vs Loss')
plt.xlabel('Learning Rate (log scale)')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()
plt.show()
```
